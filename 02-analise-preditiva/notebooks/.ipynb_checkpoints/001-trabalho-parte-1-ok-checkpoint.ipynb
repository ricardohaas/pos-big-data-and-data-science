```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


%matplotlib notebook 

```

```python
df = pd.read_csv('https://bit.ly/2Ehg4Iu')

#tratamento dos dados
#remoção de linhas com valor nulo
df2 = df.dropna()

#descrição dos dados após tratamento
#De 4240 linhas agora temos 3658
df2.describe()
```

```python
#importa a função para separação dos dados de treinamento e teste
from sklearn.model_selection import train_test_split

#pega todas as colunas
feat_list = list(df2.columns)

#remove a coluna que seria o que queremos prever
feat_list.remove('TenYearCHD')

#pega todos os indices para a entrada da previsão
X = df2[feat_list]

#pega o indice para a saída da previsão
Y = df2['TenYearCHD']

#separa os os dados de teste, deixando 75% para treinamento e 25% para teste
train_data, test_data, train_target, test_target = \
train_test_split(X, Y, train_size=0.75)
```

```python
#importa a função para regressão logística
from sklearn.linear_model import LogisticRegression
import time

#configura o método de resolução

#metodo de newton-cg para resolução, existem outros métodos possíveis
#https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
inicio = time.time()

#newton-cg ~86%, 2s
logreg = LogisticRegression(solver='newton-cg')

#sag ~85%, necessário aumentar o número de iterações para convergir, , aproximadamento 3500, ~13s
#ogreg = LogisticRegression(solver='sag',max_iter=3500)

#saga ~85%, necessário aumentar o número de iterações para convergir, aproximadamento 5000, ~22s 
#logreg = LogisticRegression(solver='saga',max_iter=5000)

#lbfgs ~86%, necessário aumentar o número de iterações para convergir, aproximadamento 2000, ~7s
#logreg = LogisticRegression(solver='lbfgs',max_iter=2000)

#liblinear ~85%, 0.5s
#logreg = LogisticRegression(solver='liblinear', verbose=1)

#treina o modelo
logreg.fit(train_data, train_target)

#neste caso o melhor modelo foi o newton-cg, que alcançou aproximandamente a mesma precisão que o lbfgs, porém executou de forma mais performática
#1.5s para 
final = time.time()
print('tempo total: {:.4f} '. format(final-inicio))

```

```python
#faz a previsao com os dados de teste
#logreg.predict(test_data)

#calcula a previsao
precisao = logreg.score(test_data, test_target)
```

```python
print('Precisão do modelo: {:.2f}%'.format(precisao*100))
```
